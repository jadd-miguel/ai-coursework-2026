```python
from tensorflow.keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
x_train, y_train = x_train[:18000], y_train[:18000]
x_test, y_test = x_test[:3000], y_test[:3000]

x_train = x_train.reshape((-1, 28, 28, 1))
x_test = x_test.reshape((-1, 28, 28, 1))
```

model1

```python
import tensorflow as tf
from tensorflow.keras import optimizers
import time
import matplotlib.pyplot as plt

model = tf.keras.Sequential([
  tf.keras.Input(shape=(28,28,1)),
  tf.keras.layers.Conv2D(filters=20,kernel_size=(3,3),activation='relu'), #(9*1+1)*20 = 200
  tf.keras.layers.MaxPooling2D(pool_size=(2,2), strides=2),
  tf.keras.layers.GlobalAveragePooling2D(),
  tf.keras.layers.Dense(10, activation='softmax')
])

optim = optimizers.Adam(learning_rate=0.01)
model.compile(optimizer=optim, loss="sparse_categorical_crossentropy", metrics=["accuracy"])

start = time.time()
hist = model.fit(
  x_train, y_train,
  epochs=8,
  batch_size=8,
  validation_data=(x_test, y_test),
  verbose=0,
)
end = time.time()
print(f"(9*1+1)*20 = 200 trainable parameters")
print(f"TRAIN TIME IS {end - start} SECONDS.")
print(f"INFERENCE TIME PER IMAGE IS {(end - start)/18000} SECONDS.")
print(f"ACCURACY IS {max(hist.history["accuracy"])}")

plt.figure()
plt.plot(hist.history["accuracy"], color="red", label="_train_loss")
plt.plot(hist.history["val_accuracy"], color="green", label="_test_loss")

plt.title('Accuracy Plot')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
```

```python
model = tf.keras.Sequential([
  tf.keras.Input(shape=(28,28,1)),
  tf.keras.layers.Conv2D(filters=5,kernel_size=(3,3),activation='relu'), #(9*1+1)*5=50
  tf.keras.layers.Conv2D(filters=8,kernel_size=(3,3),activation='relu'), #(9*5+1)*8=368, total 418
  tf.keras.layers.MaxPooling2D(pool_size=(2,2), strides=2),
  tf.keras.layers.GlobalAveragePooling2D(),
  tf.keras.layers.Dense(10, activation='softmax')
])

optim = optimizers.Adam(learning_rate=0.01)
model.compile(optimizer=optim, loss="sparse_categorical_crossentropy", metrics=["accuracy"])

start = time.time()
hist = model.fit(
  x_train, y_train,
  epochs=8,
  batch_size=8,
  validation_data=(x_test, y_test),
  verbose=0,
)
end = time.time()
print(f"(9*1+1)*5=50")
print(f"(9*5+1)*8=368, total 418")
print(f"TRAIN TIME IS {end - start} SECONDS.")
print(f"INFERENCE TIME PER IMAGE IS {(end - start)/18000} SECONDS.")
print(f"ACCURACY IS {max(hist.history["accuracy"])}")

plt.figure()
plt.plot(hist.history["accuracy"], linestyle='dotted', color="red", label="_train_loss")
plt.plot(hist.history["val_accuracy"], linestyle='solid', color="green", label="_test_loss")

plt.title('Accuracy Plot')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
```

```python
model = tf.keras.Sequential([
  tf.keras.Input(shape=(28,28,1)),
  tf.keras.layers.Conv2D(filters=5,kernel_size=(3,3),activation='relu'), #(9*1+1)*5=50
  tf.keras.layers.Conv2D(filters=8,kernel_size=(3,3),activation='relu'), #(9*5+1)*8=368, total 418
  tf.keras.layers.MaxPooling2D(pool_size=(2,2), strides=2),
  tf.keras.layers.Conv2D(filters=16,kernel_size=(3,3),activation='relu'), #(9*8+1)*16=1168, total 1636
  tf.keras.layers.GlobalAveragePooling2D(),
  tf.keras.layers.Dense(32, activation='relu'),
  tf.keras.layers.Dense(10, activation='softmax')
])

optim = optimizers.Adam(learning_rate=0.01)
model.compile(optimizer=optim, loss="sparse_categorical_crossentropy", metrics=["accuracy"])

start = time.time()
hist = model.fit(
  x_train, y_train,
  epochs=8,
  batch_size=8,
  validation_data=(x_test, y_test),
  verbose=0,
)
end = time.time()
print(f"(9*1+1)*5=50")
print(f"(9*5+1)*8=368, total 418")
print(f"(9*8+1)*16=1168, total 1636")
print(f"TRAIN TIME IS {end - start} SECONDS.")
print(f"INFERENCE TIME PER IMAGE IS {(end - start)/18000} SECONDS.")
print(f"ACCURACY IS {max(hist.history["accuracy"])}")

plt.figure()
plt.plot(hist.history["accuracy"], linestyle='dotted', color="red", label="_train_loss")
plt.plot(hist.history["val_accuracy"], linestyle='solid', color="green", label="_test_loss")

plt.title('Accuracy Plot')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
```
